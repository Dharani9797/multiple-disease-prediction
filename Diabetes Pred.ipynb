Import Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
import plotly.express as px
```

Data collection and analysis

```python
#loading the diabetes  dataset to a pandas dataframe
diabetes_dataset= pd.read_csv('/content/diabetes.csv')
```

```python
#printing the first five rows of the dataset
diabetes_dataset.head()
```

```python
#number of rows and colomns in this dataset
diabetes_dataset.shape
```

```python
#getting the staticstical measures of the data
diabetes_dataset.describe()
```

```python
diabetes_dataset['Outcome'].value_counts()
```

```python
diabetes_dataset.groupby('Outcome').mean()
```

```python
#seperating data and labels
X=diabetes_dataset.drop(columns='Outcome',axis= 1)
Y=diabetes_dataset['Outcome']
```

```python
print(X)
```

Data standardisation

```python
scaler=StandardScaler()
```

```python
scaler.fit(X)
```

```python
standardized_data=scaler.transform(X)
```

```python
X=standardized_data
Y=diabetes_dataset['Outcome']
```

Train test split

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Training the Model

```python
classifier = svm.SVC(kernel='linear')
```

```python

#training the support vector Machine Classifier
classifier.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score

```python
#accuracy score on the training data
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy score of the training data : ', training_data_accuracy)
```

```python
# accuracy score on the test data
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score of the test data : ', test_data_accuracy)
```

```python
model = svm.SVC(kernel='linear')
# training the SVM model with training data
model.fit(X_train, Y_train)
```

```python
# accuracy score on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
print('Accuracy score of training data : ', training_data_accuracy)
# accuracy score on testing data
X_test_prediction = model.predict(X_test)
tst_data_accuracy = accuracy_score(Y_test, X_test_prediction)
print('Accuracy score of test data : ', tst_data_accuracy)
```

```python

clf=LogisticRegression()
clf.fit(X_train,Y_train)
```

```python
# accuracy score on training data
clf_train_prediction = clf.predict(X_train)
train_data_accuracy = accuracy_score(Y_train, clf_train_prediction)
print('Accuracy score of training data : ', train_data_accuracy)
# accuracy score on testing data
clf_pred=clf.predict(X_test)
log_reg_accuracy = accuracy_score(Y_test,clf_pred)
print('Accuracy score of test data : ', log_reg_accuracy )
```

```python
dc=DecisionTreeClassifier()
dc.fit(X_train,Y_train)
```

```python
# accuracy score on training data
dc_train_prediction = dc.predict(X_train)
train_data_accuracy = accuracy_score(Y_train, dc_train_prediction)
print('Accuracy score of training data : ', train_data_accuracy)
# accuracy score on testing data
dc_pred=dc.predict(X_test)
dtc_acc=accuracy_score(Y_test,dc_pred)
print('Accuracy score of test data : ', dtc_acc)
```

Making a Predictive System

```python
input_data = (4,110,92,0,0,37.6,0.191,30)

# changing the input_data to numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the array as we are predicting for one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

# standardize the input data
std_data = scaler.transform(input_data_reshaped)
print(std_data)

prediction = classifier.predict(std_data)
print(prediction)

if (prediction[0] == 0):
  print('The person is not diabetic')
else:
  print('The person is diabetic')

```

```python
models = pd.DataFrame({
    'Model' : [  'SVM Classifier', 'LogisticRegression', 'Decision Tree Classifier'],
    'Score' : [tst_data_accuracy,log_reg_accuracy, dtc_acc]
})


sorted_models = models.sort_values(by = 'Score', ascending = True)

fig = px.bar(data_frame = sorted_models, x = 'Score', y = 'Model',
       title = 'Models Comparison')

fig.show()
```
